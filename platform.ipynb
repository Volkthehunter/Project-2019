```python
import cv2
import module2
import math
import time
import matplotlib.pyplot as plt
#import Image
```

```python
impath = './testimage/'#path
imname = impath + 'testimage1.png'#png image name
imname2 = impath + 'testimage1.jpg'#jpg image name
```

```python
video_capture = cv2.VideoCapture(0)
# Run the infinite loop
filename = 'feature.txt'#edit filename here
outputfile = open(filename, 'w')
start_time = time.time()
time.sleep(1)
ret, frame = video_capture.read()
cv2.imwrite(imname2, frame)#save file as jpg
while True:
    # Read each frame
    ret, frame = video_capture.read()
    # print(frame)
    # Convert frame to grey because cascading only works with greyscale image
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Call the detect function with grey image and colored frame
    canvas = module2.detect(outputfile, start_time, gray, frame)
    # Show the image in the screen
    cv2.imshow("Video", canvas)
    # Put the condition which triggers the end of program
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
outputfile.close()
```
